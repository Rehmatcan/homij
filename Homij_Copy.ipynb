# Training & Modelling #

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
# Assuming your data is in a DataFrame called df_mean
df_training = sim_grouped.copy()
df_training
```

```python
# Assuming your data is in a DataFrame called df_training

# Features (X) - all columns except IB_mean and GR_mean
X = df_training.drop(columns=['IB_mean', 'GR_mean', 'Types'])

# Target variables (y)
y = df_training[['IB_mean', 'GR_mean']]
```

```python
# Standardize the features (Simulation_Type and any other numerical variables)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# Linear Regression

```python
# Initialize and fit the linear regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predict the target variables on the test set
y_pred_lr = lr_model.predict(X_test)

# Calculate RMSE for Linear Regression
rmse_lr = np.sqrt(mean_squared_error(y_test, y_pred_lr))
print(f'Linear Regression RMSE: {rmse_lr}')
```

# Random Forest

```python
# Initialize and fit the random forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict the target variables on the test set
y_pred_rf = rf_model.predict(X_test)

# Calculate RMSE for Random Forest
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
print(f'Random Forest RMSE: {rmse_rf}')
```

```python
# Get feature importance from the Random Forest model
feature_importance = rf_model.feature_importances_

# Create a DataFrame to view feature importance
importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importance
}).sort_values(by='Importance', ascending=False)

# Display the feature importance
print(importance_df)
```

```python
# Identify the best-performing combinations directly
best_combinations_sorted = df_training.sort_values(by=['IB_mean', 'GR_mean']).reset_index(drop=True)

# Display the top 5 combinations
print(best_combinations_sorted.head())
```

```python
# Select top 5 combinations
top_5 = best_combinations_sorted.head(5)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_5, x='Simulation_Type', y='IB_mean', hue='Temperature', dodge=False)
plt.xlabel('Simulation Type')
plt.ylabel('IB_mean (Lower is Better)')
plt.title('Top 5 Best Combinations by IB_mean and Temperature')
plt.legend(title='Temperature', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
df_overall = df_training.copy()
# Aggregate by Simulation_Type
overall_performance = df_overall.groupby('Simulation_Type').agg({
    'IB_mean': 'mean',
    'GR_mean': 'mean'
}).reset_index()

# Sort by performance
overall_performance_sorted = overall_performance.sort_values(by=['IB_mean', 'GR_mean'])
print(overall_performance_sorted)
```

```python
# Check temperature coverage
valid_simulation_types = df_training.groupby('Simulation_Type')['Temperature'].nunique()
valid_simulation_types = valid_simulation_types[valid_simulation_types == 4].index

# Filter for valid simulation types
df_valid = df_training[df_training['Simulation_Type'].isin(valid_simulation_types)]

# Aggregate by Simulation_Type again
overall_performance_valid = df_valid.groupby('Simulation_Type').agg({
    'IB_mean': 'mean',
    'GR_mean': 'mean'
}).reset_index().sort_values(by=['IB_mean', 'GR_mean'])
print(overall_performance_valid)
```

```python
# Filter for top N simulation types
top_types = overall_performance_valid['Simulation_Type'].head(3)
df_top_types = df_training[df_training['Simulation_Type'].isin(top_types)]

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_top_types, x='Temperature', y='IB_mean', hue='Simulation_Type', marker='o')
plt.title('IB_mean Trends for Best Simulation Types Across All Temperatures')
plt.ylabel('IB_mean (Lower is Better)')
plt.show()
```
